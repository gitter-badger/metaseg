```python
import numpy as np
from numba import jit
from scipy.ndimage import label
from tifffile import imread, imsave
from glob import glob
from scipy import ndimage
from skimage.measure import label
```

```python
f = open('/Users/prakash/Desktop/SEGlog_STAPLE.txt')
```

```python
seg_list = []
seg_sum = 0
obj_count = 0
line = f.readline()
while True:
    # read line
    line = f.readline()
    if("T=" in line):
        seg_score = seg_sum/(obj_count+1e-10)
        seg_sum = 0
        obj_count = 0
        seg_list.append(seg_score)
    if("J=" in line):
        score = line.split("J=")[-1]
        seg_sum+= float(score)
        obj_count+=1
    # check if line is not empty
    if not line:
        seg_score = seg_sum/(obj_count+1e-10)
        seg_list.append(seg_score)
        break
f.close() 
seg_list = seg_list[1:]
```
