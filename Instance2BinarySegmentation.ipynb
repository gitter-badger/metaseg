```python
import numpy as np
import matplotlib.pyplot as plt
from tifffile import imread
from skimage import io
```

```python
def convert_instance_mask_to_binary_mask(instance_image):
    for i in range(instance_image.shape[0]):
        for j in range(instance_image.shape[1]):
            if(instance_image[i][j]==0):
                continue
            else:
                instance_image[i][j]=255
    return instance_image
```

```python
import glob, os
os.chdir("/Users/prakash/Desktop/MetaSeg_Data_Version1/TRAGen/filtered_labels/unsynched/label4/")
output_folder_path = "/Users/prakash/Desktop/MetaSeg_Data_Version1/TRAGen/filtered_binaries/binary4/"
if not os.path.exists(output_folder_path):
    os.makedirs(output_folder_path)
for file in glob.glob("*.tif"):
    instance_segmented_image = imread(file)
    fig = plt.figure()
    plt.subplot(121)
    plt.imshow(instance_segmented_image, cmap ='gray')
    plt.subplot(122)
    binary_segmented_image = convert_instance_mask_to_binary_mask(instance_segmented_image)
    plt.imshow(binary_segmented_image, cmap ='gray')
    fig.tight_layout()
    plt.show()
    io.imsave(output_folder_path+str(file), binary_segmented_image)
```
